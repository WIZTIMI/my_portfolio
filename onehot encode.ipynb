```python
x = df.drop("Survived", axis = 1)
```

```python
y = df.Survived
```

```python
x.columns = x.columns.astype(str)
x_columns = x.columns
from sklearn.preprocessing import MinMaxScaler
###this converts data to zeros and 1
scaler = MinMaxScaler()
X = scaler.fit_transform(x)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y, random_state=50)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
from sklearn.metrics import f1_score
```

```python
knn = KNN(n_neighbors=9)
```

```python
knn.fit(x_train,y_train)
```

```python
predictes_values = knn.predict(x_test)
```

```python
accuracy = f1_score(predictes_values,y_test)
```

```python
accuracy
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import mean_squared_error as MSE
```

```python
MSE(predictes_values,y_test)
```

```python
accuracy_list =[]
k_values =[]
for i in range(1,100):
    lnn = KNN(n_neighbors=i)
    lnn.fit(x_train,y_train)
    predicted_output = lnn.predict(x_test)
    accuracy = f1_score(predicted_output,y_test)
    k_values.append(i)
    accuracy_list.append(accuracy)
    
    
```

```python
plt.plot(k_values, accuracy_list)
plt.show()
##as you will see below it is 85 that is the best nneighbors
```
