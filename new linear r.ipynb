```python
def remove_outliers(df):
    for i in df.columns:
                         Q1 = df[i].quantile(0.25)
                         Q3 = df[i].quantile(0.75)
                         IQR = Q3-Q1
                         lower_bound = Q1 - 1.5 * IQR
                         upper_bound = Q3 + 1.5 * IQR
                         df_cleaned = df[(df[i] >=lower_bound) & (df[i] <= upper_bound)]
    return df_cleaned
                         
```

```python
df =remove_outliers(df)
```
