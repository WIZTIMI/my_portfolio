```python
## joining the folder together with filename
## read the image and convert it to grayscale
#resize the image
## flatten the image(1d array)
## push the flatten image into a list
## push the label into a list
```

```python
def load_imagedate(folder,label,img_size=(65,65)):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder,filename)
        img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        img_resize = cv2.resize(img,img_size)
        img_flattened = img_resize.flatten()
        images.append(img_flattened)
        labels.append(label)
    return images, labels
        
    
```

```python
cat_images, cat_labels = load_imagedate(cat_folder,0)
```

```python
dog_images, dog_labels = load_imagedate(dog_folder,1)
```

```python
X = np.array(cat_images + dog_images)
Y = np.array(cat_labels + dog_labels)
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test, y_train, y_test= train_test_split(X,Y, random_state=10, test_size = 0.2) ##test size chane the default from 70/30 to 80/20
```

```python
from sklearn.svm import SVC
```

```python
#initializing your model
svm = SVC(random_state =20)
```

```python
X.shape
```

```python
svm.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
predicted_values = svm.predict(x_test)
```

```python
print(classification_report(y_test,predicted_values))
```

```python
def predict_image(img_path):
    img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
    img_resize = cv2.resize(img,(65,65))
    img_flatten= img_resize.flatten().reshape(1,-1)
    predicted_class = svm.predict(img_flatten)
    if predicted_class == 0:
        return "CAT"
    else: return "Dog"
    return predicted_class
```
