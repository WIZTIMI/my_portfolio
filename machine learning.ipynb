### K NEAREST NEIGHBOR(KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# for your datats you need to have bot training set and test
#70% for training
#30% for test
```

```python
# x = 70% by 30%
# y = 70% by 30%
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , random_state = 20)
```

```python
x.shape
```

```python
x_train.shape
```

```python
# ;initializing the knn classifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn
```

```python
#fit the model on training set
knn.fit(x_train,y_train)
```

```python
# Accuracy score
knn.score(x_test,y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

```python
import numpy as np
```

```python
training_accuracy =[]
n_neighbor =[]
test_accuracy =[]
range(10)
```

```python
for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train,y_train)
    a = knn.score(x_test,y_test)
    print(a)
```

```python
## push the values in side the list
```

```python
for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train,y_train)
    b = knn.score(x_train, y_train)
    a = knn.score(x_test,y_test)
    training_accuracy.append(b)
    test_accuracy.append(a)
    n_neighbor.append(i)
```

```python
training_accuracy
```

```python
plt.plot(n_neighbor,training_accuracy,"-ro", color = "red")
plt.plot(n_neighbor,test_accuracy, "-ro", color = "blue", )
plt.xlabel("NUMBER OF k neighbor")
plt.ylabel("Accuracy score")
plt.xticks(np.arange(1,20))
plt.show()
### be concerned about your test
```
