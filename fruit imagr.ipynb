```python
def loadimage(folder,label,img_size=(65,65)):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder,filename)
        img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        img_resize = cv2.resize(img,img_size)
        img_flattened = img_resize.flatten()
        images.append(img_flattened)
        labels.append(label)
    return images, labels
        
    
```

```python
Pineapple_images, Pineapple_labels = loadimage(Pineapple_folder,0)
```

```python
Pineapplemini_images, Pineapplemini_labels = loadimage(Pineapplemini_folder,1)
```

```python
Raspberry_images, Raspberry_labels = loadimage(Raspberry_folder, 2)
```

```python
Redcurrant_images, Redcurrant_labels = loadimage(Redcurrant_folder, 3)
```

```python
Strawberry_images, Strawberry_labels = loadimage(Strawberry_folder, 4)
```

```python
Strawberrywedge_images, Strawberrywedge_labels = loadimage(Strawberrywedge_folder, 5)
```

```python
X = np.array(Pineapple_images + Pineapplemini_images + Raspberry_images + Redcurrant_images + Strawberry_images + Strawberrywedge_images)
```

```python
Y = np.array(Pineapple_labels + Pineapplemini_labels + Raspberry_labels + Redcurrant_labels + Strawberry_labels + Strawberrywedge_labels)
```

```python
x_train ,x_test, y_train, y_test= train_test_split(X,Y, random_state=10, test_size = 0.2) ##test size chane the default from 70/30 to 80/20
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel = "rbf" ,random_state=20)
```

```python
svm.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
predicted_values = svm.predict(x_test)
```
